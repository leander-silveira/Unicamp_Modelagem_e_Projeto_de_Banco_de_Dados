## Questões Exemplo

### [Questão Exemplo A]  Mostre um documento qualquer da coleção restaurants

````python
result = db.restaurants.find({}).limit(1)

pprint( list(result) )
````

### [Questão Exemplo B]  Encontre o documento referente ao restaurante de nome "Kosher Island"

````python
result = db.restaurants.find({"name": "Kosher Island"})

pprint( list(result) )
````

## Questões a serem resolvidas

### [Questão 1] Mostre o nome de todos os restaurantes sem repetir. Separadamente, mostre também todos os tipos distintos de cozinhas.

````python
restaurant_names = db.restaurants.distinct("name")
pprint(restaurant_names)
````

````python
cuisines = db.restaurants.distinct("cuisine")
pprint(cuisines)
````

### [Questão 2] Mostre o nome e o tipo de cozinha de 5 restaurantes quaisquer sem exibir o campo "_id"

````python
restaurants = db.restaurants.find({}, {"_id": 0, "name": 1, "cuisine": 1}).limit(5)
pprint(list(restaurants))
````

### [Questão 3] Mostre 5 restaurantes no bairro "Bronx"

````python
bronx_restaurants = db.restaurants.find({"borough": "Bronx"}).limit(5)
pprint(list(bronx_restaurants))
````

### [Questão 4] Altere o tipo de cozinha do restaurante "Kosher Island" para "Sea food"

````python
db.restaurants.update_one({"name": "Kosher Island"}, {"$set": {"cuisine": "Sea food"}})
````

### [Questão 5] Mostre e remova o(s) restaurante(s) com zipcode "10314"

````python
restaurants_10314 = db.restaurants.find({"address.zipcode": "10314"})
pprint(list(restaurants_10314))
````

````python
db.restaurants.delete_many({"address.zipcode": "10314"})
````

### [Questão 6] Adicione um campo status ao restaurante cujo nome é "Riviera Caterer". Status="demolido".

````python
db.restaurants.update_one({"name": "Riviera Caterer"}, {"$set": {"status": "demolido"}})
````

### [Questão Extra] Adicione um restaurante seguindo a estrutura de documentos da coleção

````python
novo_restaurante = {
    "address": {
        "building": "1001",
        "coord": [-73.965, 40.768],
        "street": "Broadway",
        "zipcode": "10462"
    },
    "borough": "Manhattan",
    "cuisine": "American",
    "grades": [],
    "name": "New Restaurant",
    "restaurant_id": "50075445"
}
db.restaurants.insert_one(novo_restaurante)
````

### [Questão Extra] Adicione um campo cidade do restaurante criado anteriormente para "Ilheus"

````python
db.restaurants.update_one({"name": "New Restaurant"}, {"$set": {"city": "Ilheus"}})
````

### [Questão Extra] Mostre o nome de todos os restaurantes que receberam pelo menos uma avaliação A

````python
restaurants_with_a = db.restaurants.find({"grades.grade": "A"}, {"name": 1, "_id": 0})
pprint(list(restaurants_with_a))
````
