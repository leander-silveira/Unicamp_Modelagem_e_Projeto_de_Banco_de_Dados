## Questões a serem resolvidas

### [Questão 1] Mostre o nome de todos os restaurantes sem repetir. Separadamente, mostre também todos os tipos distintos de cozinhas.

````python
# Mostrar o nome de todos os restaurantes sem repetir
restaurantes_unicos = db.restaurants.distinct("name")
print(f"Total de restaurantes únicos: {len(restaurantes_unicos)}")
for nome in restaurantes_unicos:
    print(nome)
````

````python
# Mostrando todos os tipos distintos de cozinhas
tipos_de_cozinhas = db.restaurants.distinct("cuisine")
print(f"\nTotal de tipos distintos de cozinhas: {len(tipos_de_cozinhas)}")
for cozinha in tipos_de_cozinhas:
    print(cozinha)
````

### [Questão 2] Mostre o nome e o tipo de cozinha de 5 restaurantes quaisquer sem exibir o campo "_id"

````python
# Mostrar o nome e o tipo de cozinha de 5 restaurantes quaisquer sem exibir o campo "_id"
restaurantes_qualquer = db.restaurants.find({}, {"name": 1, "cuisine": 1, "_id": 0}).limit(5)
for restaurante in restaurantes_qualquer:
    print(restaurante)
````

### [Questão 3] Mostre 5 restaurantes no bairro "Bronx"

````python
# Mostrar 5 restaurantes no bairro "Bronx"
restaurantes_bronx = db.restaurants.find({"borough": "Bronx"}, {"name": 1, "cuisine": 1, "_id": 0}).limit(5)
for restaurante in restaurantes_bronx:
    print(restaurante)
````

### [Questão 4] Altere o tipo de cozinha do restaurante "Kosher Island" para "Sea food"

````python
# Alterar o tipo de cozinha do restaurante "Kosher Island" para "Sea food"
resultado = db.restaurants.update_one(
    {"name": "Kosher Island"},
    {"$set": {"cuisine": "Sea food"}}
)

# Verificar se a alteração foi realizada com sucesso
restaurante_atualizado = db.restaurants.find_one({"name": "Kosher Island"}, {"name": 1, "cuisine": 1, "_id": 0})
print(restaurante_atualizado)
````

### [Questão 5] Mostre e remova o(s) restaurante(s) com zipcode "10314"

````python
# Mostrar o(s) restaurante(s) com zipcode "10314"
restaurantes_zipcode = db.restaurants.find({"address.zipcode": "10314"}, {"name": 1, "address": 1, "_id": 0})
for restaurante in restaurantes_zipcode:
    print("Restaurante encontrado:", restaurante)


````

````python
# Remover o(s) restaurante(s) com zipcode "10314"
resultado_remocao = db.restaurants.delete_many({"address.zipcode": "10314"})
print(f"Total de restaurantes removidos: {resultado_remocao.deleted_count}")
````

### [Questão 6] Adicione um campo status ao restaurante cujo nome é "Riviera Caterer". Status="demolido".

````python
# Adicionar um campo status ao restaurante cujo nome é "Riviera Caterer" com o valor "demolido"
resultado = db.restaurants.update_one(
    {"name": "Riviera Caterer"},
    {"$set": {"status": "demolido"}}
)

# Verificar se a alteração foi realizada com sucesso
restaurante_atualizado = db.restaurants.find_one({"name": "Riviera Caterer"}, {"name": 1, "status": 1, "_id": 0})
print(restaurante_atualizado)
````

### [Questão Extra] Adicione um restaurante seguindo a estrutura de documentos da coleção

````python
# Estrutura básica do documento de um restaurante
novo_restaurante = {
    "address": {
        "building": "1008",
        "coord": [-73.856077, 40.848447],
        "street": "Morris Park Ave",
        "zipcode": "10462"
    },
    "borough": "Bronx",
    "cuisine": "Italian",
    "grades": [
        {"date": {"$date": 1393804800000}, "grade": "A", "score": 10},
        {"date": {"$date": 1378857600000}, "grade": "B", "score": 15}
    ],
    "name": "La Dolce Vita",
    "restaurant_id": "40356018"
}

# Adicionando o novo restaurante à coleção
resultado_insercao = db.restaurants.insert_one(novo_restaurante)

# Verificar se a inserção foi realizada com sucesso
print(f"Restaurante adicionado com ID: {resultado_insercao.inserted_id}")
````

### [Questão Extra] Adicione um campo cidade do restaurante criado anteriormente para "Ilheus"

````python
# Adicionar um campo cidade ao restaurante criado anteriormente com o valor "Ilhéus"
resultado = db.restaurants.update_one(
    {"restaurant_id": "40356018"},
    {"$set": {"address.cidade": "Ilhéus"}}
)

# Verificar se a alteração foi realizada com sucesso
restaurante_atualizado = db.restaurants.find_one({"restaurant_id": "40356018"}, {"name": 1, "address": 1, "_id": 0})
print(restaurante_atualizado)
````

### [Questão Extra] Mostre o nome de todos os restaurantes que receberam pelo menos uma avaliação A

````python
# Mostrar o nome de todos os restaurantes que receberam pelo menos uma avaliação "A" em ordem alfabética
restaurantes_com_avaliacao_A = db.restaurants.find({"grades.grade": "A"}, {"name": 1, "_id": 0}).sort("name", 1)

# Exibindo o nome dos restaurantes que atenderam ao critério
for restaurante in restaurantes_com_avaliacao_A:
    print(restaurante)

````
