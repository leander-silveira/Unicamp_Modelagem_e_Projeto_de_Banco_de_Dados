# Movielens

Este [dataset](https://grouplens.org/datasets/movielens/100k/) descreve as avaliações de filmes feitas por usuários no site [MovieLens](https://movielens.org/), um serviço de recomendação de filmes. As avaliações (ratings) variam de 1 a 5, cada usuário avaliou pelo menos 20 filmes e existem simples informações demográficas sobre os usuários. Existem no banco de dados, aproximadamente, 100,000 avaliações de 1000 usuários feitas a 1700 filmes.

## Descrição das tabelas

* **Occupations**: Armazena a lista de ocupações dos usuários;
* **Users**: Armazena informações sobre os usuários, por exemplo, idade e cep;
* **Ratings**: Armazena as notas que os usuários atribuíram aos filmes e em qual momento esta nota foi atribuída;
* **Movies**: Armazena informações dos filmes;
* **Genres_movies**: Armazena o relacionamento dos gêneros e dos filmes;
* **Genres**: Armazena a lista de gêneros.

## [Questão 15] Crie uma View com a média de avaliações de cada filme. A partir da view, quais os nomes dos filmes com maior média de avaliação? 

<!--

```sql
DROP VIEW IF EXISTS movie_average_score;

CREATE VIEW movie_average_score AS 
SELECT
  m.id, 
  m.title, 
  AVG(r.rating) AS average_score,
  COUNT(r.rating) AS rating_count
FROM 
  movies AS m 
  JOIN ratings AS r ON m.id = r.movie_id 
GROUP BY 
  m.id, m.title;

SELECT 
  title, 
  average_score
FROM 
  movie_average_score 
ORDER BY 
  average_score DESC 
LIMIT 20;
```

## [Desafio - Parte 1] Como ficaria o modelo lógico do banco de dados para que fosse possível implementar a seguinte funcionalidade: "usuários podem ser amigos de outros usuários"?


```sql
/*
Assumindo que amizades sâo bilaterais, ou seja, se A é amigo de B, o inverso também é verdadeiro
nao se torna necessario adicionar duas tuplas Ex. (1, 2) e (2, 1) para representar uma amizade entre dois usuários,
ou seja, garantir integridade a longo prazo exige que: para cada celebração de amizade haverá apenas uma tupla na relação "friendship"
*/
CREATE TABLE IF NOT EXISTS friendship (
  user1_id int(11) NOT NULL, 
  user2_id int(11) NOT NULL, 
  PRIMARY KEY (user1_id, user2_id), 
  CONSTRAINT friendship_users_FK_1 FOREIGN KEY (user1_id) REFERENCES users (id), 
  CONSTRAINT friendship_users_FK_2 FOREIGN KEY (user2_id) REFERENCES users (id), 
  CONSTRAINT self_friendship_CHECK CHECK (user1_id <> user2_id)
);

-- π x,y(R) ∩ π y/x,x/y(R) = ∅
DELIMITER ;;
CREATE TRIGGER IF NOT EXISTS prevent_inverted_friendship_duplicates BEFORE INSERT ON friendship
  FOR EACH ROW BEGIN
    IF EXISTS
      (SELECT 1 FROM friendship WHERE user1_id = new.user2_id AND user2_id = new.user1_id)
    THEN
      SIGNAL SQLSTATE '23000' SET message_text = 'Friendship already in place on reverse order.';
    END IF;
  END;;
DELIMITER ;

/*
π x,y(R) ∪ π y/x,x/y(R)
Visualizar os relacionamentos de amizade nos dois sentidos
Esta view é útil para resolver a parte 3 do desafio
*/
CREATE VIEW IF NOT EXISTS two_way_friendship_view AS 
SELECT 
  user1_id, 
  user2_id 
FROM 
  friendship 
UNION ALL 
SELECT 
  user2_id AS user1_id, 
  user1_id AS user2_id 
FROM 
  friendship;
```

## [Desafio Parte 3] Quais são os filmes avaliados pelos amigos dos amigos do usuário de id 1? 

Se for necessário, crie comandos SQL para popular o seu banco de dados para adicionar amigos ao usuário de id 1 e outros usuários

```sql
-- Considerando que o usuário 1 também é amigo de seus amigos
WITH friends_of_friends AS (
  SELECT DISTINCT user2_id
  FROM two_way_friendship_view
  WHERE user1_id IN (
    SELECT user2_id
    FROM two_way_friendship_view
    WHERE user1_id = 1
  )
)
SELECT DISTINCT m.title 
FROM users u 
JOIN ratings r ON u.id = r.user_id 
JOIN movies m ON m.id = r.movie_id 
WHERE u.id IN (SELECT user2_id FROM friends_of_friends)
LIMIT 50;
```
