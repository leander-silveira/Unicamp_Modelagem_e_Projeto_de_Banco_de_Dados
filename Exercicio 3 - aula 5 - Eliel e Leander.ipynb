````python
%%cypher

MATCH (node:Movie)
WHERE node.title='Cloud Atlas'
RETURN node
````

## [Questão Exemplo]  Liste apenas o nome de 10 filmes

````python
%%cypher

MATCH (node:Movie)
RETURN node.title LIMIT 10
````

## [Questão Exemplo] Liste todos os filmes em que Tom Hanks atuou

````python
%%cypher

MATCH (node)-[:ACTED_IN]->(node2:Movie)
WHERE node.name='Tom Hanks'
RETURN node2.title as Titulo_do_Filme
````

## [Questão 1] Liste o nome de até 10 filmes lançados entre 1990 e 1999

````python
%%cypher

MATCH (m:Movie)
WHERE m.released >= 1990 AND m.released <= 1999
RETURN m.title AS movie_title
LIMIT 10;

````

## [Questão 2] Qual(is) o(s) nome(s) do(s) diretor(es) do filme Cloud Atlas?

````python
%%cypher

MATCH (m:Movie {title: "Cloud Atlas"})<-[:DIRECTED]-(d:Person)
RETURN d.name AS director_name;

````

## [Questão 3] Quais são os nomes do outros atores que atuaram no filme Cloud Atlas?

````python
%%cypher

MATCH (m:Movie {title: "Cloud Atlas"})<-[:ACTED_IN]-(a:Person)
WHERE a.name <> "Tom Hanks"
RETURN a.name AS actor_name;

````

## [Questão 4] Existe algum filme em que o Tom Hanks tenha dirigido e atuado? Se sim, qual o nome e ano de lançamento?

````python
%%cypher

MATCH (t:Person {name: "Tom Hanks"})-[:ACTED_IN]->(m:Movie)<-[:DIRECTED]-(t)
RETURN m.title AS movie_title, m.released AS release_year;
````

## [Questão 5] Liste o nome de 10 diretores que dirigiram filmes em que o Tom Hanks atuou

dica:
``` cypher
MATCH (a)-[:ACTED_IN]->(m)<-[:DIRECTED]-(d) RETURN a,m,d LIMIT 10
```                                                                           

````python
%%cypher

MATCH (t:Person {name: "Tom Hanks"})-[:ACTED_IN]->(m:Movie)<-[:DIRECTED]-(d:Person)
RETURN DISTINCT d.name AS director_name
LIMIT 10;

````

## [Questão 6] Existe algum filme em que ambos Kevin Bacon e Tom Cruise tenham atuado? Se sim, qual o nome e ano de laçamento?

````python
%%cypher

MATCH (kevin:Person {name: "Kevin Bacon"})-[:ACTED_IN]->(m:Movie)<-[:ACTED_IN]-(tom:Person {name: "Tom Cruise"})
RETURN m.title AS movie_title, m.released AS release_year;

````

## [Questão 7] Liste o nome de 10 pessoas que também atuaram em filmes com o Tom Hanks

````python
%%cypher

MATCH (t:Person {name: "Tom Hanks"})-[:ACTED_IN]->(m:Movie)<-[:ACTED_IN]-(co_actor:Person)
RETURN DISTINCT co_actor.name AS co_actor_name
LIMIT 10;

````

## [Questão extra] Quem mais produziu filmes do Tom Hanks e do Kevin Bacon?

````python
%%cypher

MATCH (t:Person {name: "Tom Hanks"})-[:ACTED_IN]->(m:Movie)<-[:PRODUCED]-(p:Person),
      (k:Person {name: "Kevin Bacon"})-[:ACTED_IN]->(m)
RETURN p.name AS producer_name;

````

## [Questão Extra] Quem tem mais seguidores?

````python
%%cypher

MATCH (p:Person)-[:FOLLOWS]->(f:Person)
RETURN f.name AS person_name, COUNT(p) AS followers_count
ORDER BY followers_count DESC
LIMIT 1;

````

## [Questão Extra] Liste o(s) nomeQ(s) da(s) pessoa(s) que produziram, escreveram e dirigiram o mesmo filme

````python
%%cypher

MATCH (p:Person)-[:PRODUCED]->(m:Movie),
      (p)-[:DIRECTED]->(m),
      (p)-[:WROTE]->(m)
RETURN DISTINCT p.name AS person_name, m.title AS movie_title;

````

## [Questão Extra] Qual a média de notas (reviewed) do filme The Replacements?

````python
%%cypher

MATCH (m:Movie {title: "The Replacements"})<-[r:REVIEWED]-(p:Person)
RETURN AVG(r.rating) AS average_rating;
````
